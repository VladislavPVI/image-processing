```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("cat.jpg").convert('L')
```

```python
img
```

```python
a = np.array(img.convert('L'))
a
```

```python
plt.hist(a.flatten(),256,[0,256]) 
plt.show() 
```

```python
def threshold(intensity):
    thr = 190
    iI = intensity
    iO = 0
    if iI > thr:
        iO = 255 
    return iO
```

```python
img2 = img.point(threshold)
img2
```

```python
img = Image.open("low.png")
img
```

```python
a = np.array(img)
plt.hist(a.flatten(),256,[0,256]) 
plt.show() 
```

```python
minI    = a.min()
maxI    = a.max()
minO    = 0
maxO    = 255
minI, maxI
```

```python
def normalize(intensity):

    iI      = intensity
    iO      = ((iI - minI) / (maxI - minI)) * (maxO - minO) + minO
    return iO
```

```python
img2 = img.point(normalize)
img2
```

```python
a = np.array(img2)
plt.hist(a.flatten(),256,[0,256]) 
plt.show() 
```

```python
a.min(),a.max()
```

```python
def taskNormalize(intensity):
    iI = intensity
    left = 53
    right = 250
    if iI < left:
        return 0
    if iI > right:
        return 255
    iO = ((iI - left) / (right - left)) * (maxO - minO) + minO
    return iO
```

```python
img3 = img.point(taskNormalize)
img3
```

```python
a = np.array(img3)
plt.hist(a.flatten(),256,[0,256]) 
plt.show() 
```
